```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('train.csv')
train
```

```python
train.rating.value_counts()
```

```python
train_lama = pd.read_csv('test lama.csv')
sub_lama = pd.read_csv('Submission Lama.csv')
train_lama['rating'] = sub_lama['rating']
train_lama
```

```python
baru = pd.concat([
    train,
    train_lama[(train_lama['rating'] == 1) | (train_lama['rating'] == 2)],
    train_lama[train_lama['rating'] == 5].sample(n = 41865 - 41515),
    train_lama[train_lama['rating'] == 3].sample(n = 41865 - 35941)
])
baru.shape
```

```python
baru['review_id'] = list(range(164866))
baru
```

```python
import seaborn as sns
sns.countplot(baru.rating)
```

```python
baru.to_csv('train baru.csv')
```
