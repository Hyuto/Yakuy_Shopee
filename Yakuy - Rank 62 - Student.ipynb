```python
# Basic
import numpy as np
import pandas as pd
import datetime as dt
from collections import Counter
```

```python
# Load Dataset
data = pd.read_csv('C:/Users/Wahyu Setianto/Downloads/order_brush_order.csv')
data
```

```python
# Dataset
data['event_time'] = pd.to_datetime(data['event_time'])
data.info()
```

```python
# Sort data
data_new = data.sort_values(by=['shopid', 'event_time'])
data_new
```

```python
test = data['shopid'].value_counts()
test1 = test[test >= 3]
test2 = test[test < 3]
test1.index
```

```python
# Bellow 3 Transaction
bellow3 = pd.DataFrame({'shopid' : test2.index, 'userid' : [0] * len(test2.index)})
bellow3
```

```python
# Up 3 Transaction
res = {}
for shop in test1.index:
    res[shop] =  [] 
    temp_data = data_new[data_new['shopid'] == shop]
    for time in temp_data['event_time']:
        temp_temp_data = temp_data[(temp_data['event_time'] >= time) & (temp_data['event_time'] <= time + dt.timedelta(hours = 1))]
        score = len(temp_temp_data) / len(temp_temp_data['userid'].drop_duplicates())
        if score >= 3:
            user = temp_data['userid'].tolist()
            count = Counter(user)
            prop = sorted([(i, count[i] / len(user)) for i in count], reverse = True, key = lambda x : x[1])
            for i in range(len(prop)):
                if prop[i][1] == prop[0][1]:
                    if prop[i][0] not in res[shop]: 
                        res[shop].append(prop[i][0])
```

```python
for x in res:
    if res[x] == []:
        res[x] = 0
    else:
        res[x] = '&'.join(str(y) for y in sorted(res[x]))
```

```python
up3 = pd.DataFrame({'shopid' : list(res.keys()), 'userid' : list(res.values())})
```

```python
len(bellow3) + len(up3)
```

```python
hasil = pd.concat([bellow3, up3])
```

```python
hasil.to_csv('Yakuy_Submission.csv', index=False)
```
